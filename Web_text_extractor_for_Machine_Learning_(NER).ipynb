# **Web checker**

We check every webpage if is good for text extraction, not found or if the website is down.

```python
def web_extractor():
  list = []
  with open('/content/furniture_stores_pages.csv', newline='') as file:
      reader= csv.reader (file, delimiter=' ')
      for row in reader:
          for url in row:
            try:
                r=requests.get(url,timeout=5)
                r.encoding = "ISO-8859-1"
                if r:
                  print('Success!', url)
                  list.append(url)

                  with open("good_webs23.csv", "w") as file:
                      for element in list:      
                          file.write(element + "\n")
                else:
                  print('404 - not found.', url)
                
            except Exception as e:
              print(f"There was an error: {e}",url)
              pass

web_extractor()
```

# For multiple webpages (CSV file)

```python
import csv
import requests
from bs4 import BeautifulSoup
import itertools
import io


def text_extractor():
  data = []
  with open('/content/good_webs23.csv', newline='') as file:
        reader= csv.reader (itertools.islice(file,110), delimiter=' ')
        for row in reader:
            for url in row:
              page_status = requests.get(url)
              page_status.encoding = "ISO-8859-1"
              soup = BeautifulSoup(page_status.content, 'html.parser')
              with io.open('logtxt.txt', 'a', encoding='utf8') as logfile:
                for data in soup.find_all(["h1","h2","h3","h4"]):
                  text=data.get_text()
                  logfile.write(text.strip() + "\n")
                  # file.write(text.strip() + "\n")
                  print(text) 
    
# "h2","h3","h4","h5","span"
text_extractor()       
```

# **Concatenation**

```python
filenames = ['/content/logtxt.txt', '/content/wiki.txt']
with open('/content/concatenate_file.txt', 'w') as outfile:
    for fname in filenames:
        with open(fname) as infile:
            for line in infile:
                outfile.write(line)
```

# **Data frame (Optional)**

```python
import os
import spacy 
from spacy import displacy
from spacy.tokens import Span
from spacy.symbols import ORTH
import re
from dframcy.matcher import DframCyMatcher, DframCyPhraseMatcher, DframCyDependencyMatcher

with open('/content/concatenate_file.txt', 'r') as file:
    doc = file.read()

nlp = spacy.load("en_core_web_sm")
doc = nlp(doc)

for line in doc.ents:
  print([(line.text, line.label_)])
```

```python
from dframcy import DframCy

dframcy = DframCy(nlp)
spacy_df = dframcy.to_dataframe(doc, ['is_sent_start', 'text','pos_','tag_' ]).reset_index()
spacy_df.token_is_sent_start = spacy_df.token_is_sent_start.astype(bool).cumsum() - 1
spacy_df = spacy_df.rename(columns={'token_is_sent_start': 'Sentence #',
                                    # 'index': 'token_id',
                                    'token_text': 'Token', 'POS':'pos_','Tag':'tag_' })


spacy_df.to_csv('data_frame.csv')
print(spacy_df)
```
